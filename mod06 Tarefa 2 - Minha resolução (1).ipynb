```python
import pandas as pd

sinasc = pd.read_csv('Sinasc.csv')

print(sinasc.head())
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x=sinasc['IDADEMAE'], color='skyblue')
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=sinasc['PESO'], color='lightcoral')
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(10, 6))
sns.histplot(sinasc['IDADEMAE'], bins=30, kde=False, color='skyblue')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Contagem')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(10, 6))
sns.histplot(sinasc['PESO'], bins=30, kde=False, color='salmon')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso do Bebê')
plt.ylabel('Contagem')
plt.show()
```
