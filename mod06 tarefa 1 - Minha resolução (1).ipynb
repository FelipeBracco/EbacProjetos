#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd

sinasc = pd.read_csv('Sinasc.csv')

print(sinasc.head())
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(x='APGAR1', data=sinasc, palette='viridis')
plt.title('Distribuição de APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Contagem')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='APGAR5', data=sinasc, palette='viridis')
plt.title('Distribuição de APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Contagem')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(x='SEXO', data=sinasc, palette='viridis')
plt.title('Distribuição de SEXO')
plt.xlabel('SEXO')
plt.ylabel('Contagem')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='GESTACAO', data=sinasc, palette='viridis')
plt.title('Distribuição de GESTACAO')
plt.xlabel('GESTACAO')
plt.ylabel('Contagem')
plt.xticks(rotation=45)  
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='ESTCIVMAE', data=sinasc, palette='viridis')
plt.title('Distribuição de ESTCIVMAE')
plt.xlabel('ESTCIVMAE')
plt.ylabel('Contagem')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc, palette='viridis', errorbar=None)
plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Distribuição de Variáveis', fontsize=16)

sns.countplot(x='SEXO', data=sinasc, palette='pastel', ax=axes[0, 0])
axes[0, 0].set_title('Distribuição de SEXO')

sns.countplot(x='GESTACAO', data=sinasc, palette='pastel', ax=axes[0, 1])
axes[0, 1].set_title('Distribuição de GESTACAO')
axes[0, 1].tick_params(axis='x', rotation=45)

sns.countplot(x='ESTCIVMAE', data=sinasc, palette='pastel', ax=axes[1, 0])
axes[1, 0].set_title('Distribuição de ESTCIVMAE')
axes[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(x='CONSULTAS', data=sinasc, palette='pastel', ax=axes[1, 1])
axes[1, 1].set_title('Distribuição de CONSULTAS')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
